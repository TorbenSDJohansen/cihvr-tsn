```python
import numpy as np
import json
```

```python
make_cloud = lambda x: list([x, int(y)] for y in np.linspace(150, 280, num=10))

def to_json(name, d):
    with open(name, 'w') as f:
        json.dump(d, f, indent=4)
```

NOTE:

Template X-coordinates: {75, 153, 225}
Template Y-coordinates: Range from 150 - 280, linearly sampling 10 points for each X-coordinate


Target is slightly shifted to the right and down, to simulate real world cases.

Target X-coordinates = {85, 163, 235}
Target Y-coordinates = Range from 160 - 290, linearly sampling 10 points for each X-coordinate

```python
def make_template():
    template = {}

    template['width'] = 291
    template['height'] = 414

    template['cloud0'] = make_cloud(75)
    template['cloud1'] = make_cloud(153)
    template['cloud2'] = make_cloud(225)
    return template

def make_target():
    target = {}
    
    target['width'] = 291
    target['height'] = 414
    
    target['cloud0'] = make_cloud(85)
    target['cloud1'] = make_cloud(163)
    target['cloud2'] = make_cloud(235)
    return target
```

```python
to_json('template_mock.json', make_template())
to_json('target_mock.json', make_target())
```

```python
d = make_template()
```
